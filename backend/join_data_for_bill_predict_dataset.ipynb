```python
new_df = pd.DataFrame()
new_df.info()
```

```python
new_row = pd.DataFrame({
        "Компания": [0],	
        "Год": [0],	
        "Номер счета": [0],	
        "Позиция счета": [0],	
        "ID услуги": [0],	
        "ID договора": [""],	
        "Дата отражения счета в учетной системе": [""],	
        "Стоимость без НДС": [0],
        "Сумма площадей": [0]})
cur_dist_num = 0
with progressbar.ProgressBar(max_value=df.shape[0]) as bar:
    for _, row in df.iterrows():
        if row["Номер позиции распределения"] - cur_dist_num != 1:
            new_df = pd.concat([new_df, new_row])
            cur_dist_num = 0 
            new_row = pd.DataFrame({
                "Компания": [0],	
                "Год": [0],	
                "Номер счета": [0],	
                "Позиция счета": [0],	
                "ID услуги": [0],	
                "ID договора": [""],	
                "Дата отражения счета в учетной системе": [""],	
                "Стоимость без НДС": [0],
                "Сумма площадей": [0]})

        new_row["Стоимость без НДС"] = new_row["Стоимость без НДС"] + row["Сумма распределения"]
        new_row["Компания"] = row["Компания"]
        new_row["Год"] = row["Год счета"]
        new_row["Номер счета"] = row["Номер счета"]
        new_row["Позиция счета"] = row["Позиция счета"]
        new_row["ID услуги"] = row["Услуга"]
        new_row["ID договора"] = row["ID договора"]
        new_row["Дата отражения счета в учетной системе"] = row["Дата отражения в учетной системе"]
        new_row["Сумма площадей"] = new_row["Сумма площадей"] + row["Площадь"]
        #print(new_row)
        cur_dist_num += 1
        bar.update(_)  
new_df.to_excel("../data/example_output/xd.xlsx")
        
```
